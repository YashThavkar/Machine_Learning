```python
import pandas as pd
df = pd.read_csv('titanic.csv')
df.head()
```

```python
dir(df)
```

```python
#removing unnecessary columns from the dataset
df.drop(['Cabin','Embarked','Name','Parch','PassengerId','Pclass','SibSp','Ticket'], axis='columns', inplace=True)
```

```python
df.head()
```

```python
target =df.Survived
target.head()
```

```python
inputs =df.drop('Survived', axis ='columns')
inputs.head()
```

# converting text into dummines

```python

dummies = pd.get_dummies(inputs.Sex)
dummies.head()
# inputs.head()
```

```python
inputs = pd.concat([inputs,dummies], axis ='columns')
inputs.head()
```

```python
inputs_male=inputs.drop(['male'], axis='columns', inplace=True)
inputs.head()
```

```python
inputs.Sex = inputs.drop(['Sex'], axis ='columns', inplace =True)
inputs.head()
```

# finding out the NA values in dataset

```python
inputs.isna().any()
```

```python
inputs.isna().sum()
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs.head()
```

```python
inputs.isna().sum()
```

# Train Test Split Method

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test=train_test_split(inputs, target,test_size=0.3)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
inputs.head()
```

# Naive Baise Algorithm

```python
from sklearn.naive_bayes import GaussianNB
nb =GaussianNB()
```

```python
nb.fit(X_train,y_train)
```

```python
nb.score(X_test,y_test)
```

```python
nb.predict(X_test[:10])
```

```python
nb.predict_proba(X_test[:10])
```

```python
#the score using cross validation
from sklearn.model_selection import cross_val_score
cross_val_score(GaussianNB(),X_train, y_train, cv=5)
```
